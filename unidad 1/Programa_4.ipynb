```python
import random
edades = []
#ciclo para generar 100 numeros aleatorios (representan edades)
for i in range(100):
  #Generamos un numero aleatorio y lo guardamos en la variable randint
  numero_aleatorio = random.randint( 18,70 )
  #Agregamos el numero a la lista de valores
  edades.append(numero_aleatorio)
print(edades)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(edades, columns=['Edad'])
df.head()
```

```python
#Generar una tabla de distribucion de frecuencia
tabla_frecuencias = df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad', 'Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by='Edad')

#Imprimir la tabla
print("Tabla de Distribución de Frecuencia:")
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Relativa'] = tabla_frecuencias['Frecuencia']/100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Porcentual'] = tabla_frecuencias['Frecuencia'] /100 *100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Acumulada'] = tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

# Regla de sturges
1.- Es una regla que sirve para calcular el número de clases o intérvalos idóneo en los que se de dividir un conjunto de datos

2.- Establece que el **Número de clases es igual a uno más el logaritmo en base dos del número total de datos**


```python
import math
```

```python
print(edades)
```

```python
df = pd.DataFrame({'Edad': edades})
df
```

```python
n = len(edades)
K = int( 1 + math.log2(100))
print("Valor de n = ", n , "\nNúmero de intervalos =", K)
```

```python
tabla_frecuencias = pd.cut(df['Edad'], bins=K, include_lowest=True).value_counts().reset_index()
tabla_frecuencias.columns = ['Clase', 'Frec']
tabla_frecuencias
```

```python
tabla_frecuencias = tabla_frecuencias.sort_values(by='Clase')
tabla_frecuencias
```

```python
#Crear columnas
tabla_frecuencias['Frec Rel'] = tabla_frecuencias['Frec'] / n
tabla_frecuencias['Frec Porc'] = tabla_frecuencias['Frec'] /n * 100
tabla_frecuencias['Frec Ac'] = tabla_frecuencias['Frec'].cumsum()
tabla_frecuencias['Frec Rel Ac'] = tabla_frecuencias['Frec Rel'].cumsum()
tabla_frecuencias['Frec Porc Ac'] = tabla_frecuencias['Frec Porc'].cumsum()

#Mostrar la tabla de frecuencias
print(tabla_frecuencias)
```

```python
# Crear un histograma en la tabla de frecuencia
plt.hist(df['Edad'], bins = K, alpha = 0.5)
plt.title("Histograma de edades basados en la regla de sturges")
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
```
